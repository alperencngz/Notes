```python
# Şimdi elbow plot yöntemiyle optimum cluster sayısını bulacağım
```

```python
from sklearn.cluster import KMeans
```

```python
# elbowplota girilecek grafik verilerini üretiyoruz
def optimise_k_means(data, max_k):
    means = []
    inertias = []
    
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters = k)
        kmeans.fit(data)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
        
    # grafiğin çizimi
    fig = plt.subplots(figsize = (10,5))
    plt.plot(means, inertias, "o-")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Inertia")
    plt.grid(True)
    plt.show()
```

```python
optimise_k_means(X, 10)
```

##### Grafiğe baktığımızda optimum sayı 4 ya da 5 gibi duruyor, verilerimizin çokluğunu düşünerek 5 gruba ayıracağım.
